```python
for col in ['Sales', 'Profit']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = ((df[col] < lower) | (df[col] > upper)).sum()
    print(f"{col}: {outliers} outliers out of {len(df)} rows")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")

df = pd.read_csv("/content/Sample - Superstore.csv", encoding='latin1')
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
plt.figure(figsize=(8,5))
df.groupby('Category')['Sales'].sum().sort_values(ascending=False).plot(kind='bar', color='steelblue')
plt.title("Total Sales by Category")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Category', y='Sales', data=df)
plt.title("Sales Spread by Category")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='Category', y='Sales', data=df)
plt.yscale('log')
plt.title("Sales Spread by Category (Log Scale)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

## Key Insights

1. **Sales Distribution**: Sales is heavily right-skewed â most orders are under â¹500, with a long tail of large bulk orders reaching over â¹22,000.

2. **Sales by Category**: Technology generates the highest total sales (~â¹836K), followed closely by Furniture (~â¹742K) and Office Supplies (~â¹720K). Sales are fairly evenly distributed across categories.

3. **Outliers**: Sales and Profit have a significant number of outliers (11.7% and 18.8% respectively), but these largely represent legitimate large transactions and loss-making discounted orders â so they were retained rather than removed.

4. **Sales Spread by Category**: Technology and Furniture show similar median sales and wider spread, while Office Supplies has a lower median, consistent with it being made up of many low-cost items.

5. **Correlations**: Sales and Profit show a moderate positive correlation (0.48). Discount is negatively correlated with Profit (-0.22), confirming heavy discounting hurts margins. Quantity has only a weak relationship with Profit (0.07), meaning profit is driven more by price/discount than order volume.
